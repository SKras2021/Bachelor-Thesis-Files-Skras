```python
#imports
import os
import json
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
"TRIAL 1"
```

```python
dir = os.path.join(os.getcwd(), "temp_saves")
print(dir)

df = []
```

```python

# Traverse each JSON file in the folder

for i in range(1,301,1):
    filepath = os.path.join(dir, f"save2_{i}.json")
    with open(filepath, 'r') as f:
        data = json.load(f) 
        df.append(data)

print(f"loaded {len(df)} files")
```

```python
df = pd.DataFrame(df)
df.shape
```

```python
df.head()
```

```python
#here we see history of each tile by turn. 1620 tiles in total over the course of 300 turns in our case
```

```python
#Only looking at whether the creature is alive or not
alive = df.map(lambda x: x[0] if isinstance(x, list) and len(x) > 0 else 0)
alive
```

```python
row_sums = alive.sum(axis=1)
row_sums
```

```python
plt.plot(row_sums)
plt.xlabel("Time (Turns)")
plt.ylabel("Population")
plt.title("Number of creatures alive (total space 1620)")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
#temp saves 300, temp saves 2 2500
dir_name = "temp_saves2"
length_simulation = 2500
dir = os.path.join(os.getcwd(), dir_name)
print(dir)

df = []

for i in range(1,length_simulation+1,1):
    filepath = os.path.join(dir, f"save2_{i}.json")
    with open(filepath, 'r') as f:
        data = json.load(f) 
        df.append(data)

print(f"loaded {len(df)} files")

df = pd.DataFrame(df)
alive = df.map(lambda x: x[0] if isinstance(x, list) and len(x) > 0 else 0)
row_sums = alive.sum(axis=1)

lst = ["Population count", "genome complexity", "average agressivness", "stochasticity", " average fertility", "average rigidness", "average speed", "energy costs", "enegy_reseve","survivability", "apaptability", "stporage complex", "dynamics complex", "impactfullness", "development"]
row_sums = []
for i in range(15):
    x = df.map(lambda x: x[i] if isinstance(x, list) and len(x) > 0 else 0)
    if (i != 0):
        x = x.sum(axis=1) / row_sums
    else:
        row_sums = x.sum(axis=1)
        x = x.sum(axis=1)
        
    plt.plot(x)
    plt.xlabel("Time (Turns)")
    plt.ylabel(f"Char {lst[i]}")
    plt.title(f"{lst[i]} overtime")
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```
